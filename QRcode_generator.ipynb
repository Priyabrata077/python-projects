```python
import qrcode
import image
```

```python
qr = qrcode.QRCode(
    version = 15,
    box_size=10,
    border = 5
)
```

```python
data = "https://github.com/Priyabrata077"

qr.add_data(data)
qr.make(fit=True)

img = qr.make_image(fill_color="black", back_color = "white")
img.save('github.png')
```
